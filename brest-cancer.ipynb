```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("data.csv")
dataset.head(3)
```

```python
dataset = dataset.drop(columns='Unnamed: 32' , axis=1)
```

```python
dataset = dataset.drop(columns='id' , axis=1)
```

```python
dataset.head(3)
```

```python
mapping ={'M':0 , 'B':1}
dataset['diagnosis'] = dataset['diagnosis'].map(mapping)
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated().sum()
```

```python
dataset.describe()
```

```python
x = dataset.drop(columns='diagnosis' , axis=1)
y = dataset['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm= MinMaxScaler()
```

```python
x_train = mm.fit_transform(X_train)
x_test=mm.transform(X_test)
```

```python
x_train.shape  , x_test.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
ann = keras.Sequential([
    keras.layers.Dense(64, input_shape = (30 ,) , activation=('relu')),
    keras.layers.Dense(32 , activation=('relu')),
    keras.layers.Dense(16 , activation=('relu')),
    keras.layers.Dense(1 , activation=('sigmoid'))
])
ann.summary()
    
```

```python
ann.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics=['accuracy'])
```

```python
ann.fit(x_train, y_train , epochs=100)
```

```python
ann.evaluate(x_test , y_test)
```
